# Analisis univariado

Tomando como variable respuesta la velocidad del viento horaria, realizamos una analisis especifico al comportamiento de la variable.

```python
import matplotlib.pyplot as plt 
import seaborn as sns
print('Column: Velocidad horario del viento (m/s)')
print('Skew:', round(wind_speed['Velocidad horario del viento (m/s)'].skew(), 2))
print('Kurtosis: ', round(wind_speed['Velocidad horario del viento (m/s)'].kurtosis(), 2))
plt.figure(figsize = (14, 6))
plt.subplot(1, 2, 1)
#wind_speed['Velocidad horario del viento (m/s)'].hist(grid=False)
sns.distplot(wind_speed['Velocidad horario del viento (m/s)'], axlabel="Velocidad horario del viento");
plt.title('Historgrama de Velocidad horario del viento')
plt.subplot(1, 2, 2)
sns.boxplot(wind_speed['Velocidad horario del viento (m/s)'])
plt.title('Boxplot de Velocidad horario del viento')
plt.show()
```

Basado en los resultados podemos identificar que la variable 'velocidad horario del viento':

**Asimetria:**
(Skew: 0.37) Presenta un ligero sesgo positivo, esto quiere decir que tiene mayor peso en la cola izquierda de la distribucion, tal y como se obserba en el histograma.

**Kurtosis:**
(Kurtosis:  -0.11) La forma de la distribucion de frecuencias es platicurtica, esto significa que la distribucion presenta colas menos gruesas que la normal.

## Modelo KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
wind_speed_X = wind_speed.drop(['HORA (UTC)',
                     'Humedad rel. max. hora anterior (AUT)(%)',
                     'Humedad rel. min. hora anterior (AUT)(%)',
                     'Presion atmosferica max. hora anterior (AUT)(mB)',
                     'Presion atmosferica min. hora anterior (AUT)(mB)',
                     'Maxima rafaga de viento (m/s)',
                     'Temperatura min. hora anterior (AUT)',
                     'Velocidad horario del viento (m/s)'
                     
                     ],axis=1)
wind_speed_y = wind_speed['Velocidad horario del viento (m/s)']
```

```python
from sklearn.model_selection import train_test_split

X = wind_speed_X
y = wind_speed_y
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)
```

```python
print("Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))
```

Tomando set de datos elimando las variables altamente correlacionadas el coefeciente de determinacion $\R^2$ fue del 0.10, por lo tanto se tomara el dataset completo. 

```python
wind_speed_X = wind_speed.drop(['HORA (UTC)',
                                'Velocidad horario del viento (m/s)'              
                               ],axis=1)
wind_speed_y = wind_speed['Velocidad horario del viento (m/s)']
```

```python
from sklearn.model_selection import train_test_split

X = wind_speed_X
y = wind_speed_y
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)
```

```python
print("Test set predictions:\n{}".format(reg.predict(X_test)))
```

```python
print("Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))
```

```python
y_pred = reg.predict(X_test)
```

### MAPE

```python
import numpy as np
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    # Evitar división por cero
    mask = y_true != 0
    return np.mean(np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])) * 100

# Calcular MAPE
mape = mean_absolute_percentage_error(y_test, y_pred)
print("MAPE: %f" % (mape))
```

### RMSE

```python
from sklearn.metrics import mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE: %f" % (rmse))
```
